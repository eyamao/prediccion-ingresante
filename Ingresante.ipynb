### Prediccion

```python
#train test split
from sklearn.model_selection import train_test_split
y = data["APROBADO"]
X = data[["EDAD","SEXO","COLEEXC","DISTANCIA","EXADMISION"]]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

#### Logistic Regression

```python
log_reg = sm.Logit(y_train, X_train).fit()
```

```python
print(log_reg.summary())
```

```python
yhat = log_reg.predict(X_test)
prediction = list(map(round, yhat))
```

```python
from sklearn.metrics import (confusion_matrix, accuracy_score)

# confusion matrix
cm = confusion_matrix(y_test, prediction)
print ("Confusion Matrix : \n", cm)

# accuracy score of the model
print('Test accuracy = ', accuracy_score(y_test, prediction))
```

#### Arbol de Decisiones

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
from sklearn.model_selection import train_test_split
y = data["APROBADO"]
#X = data[["EDAD","SEXO","COLEEXC","DISTANCIA","EXADMISION","PROVINCIA","COLPRE","COLPRIV"]]
X = data[["EDAD","SEXO","EXADMISION"]]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
clf = DecisionTreeClassifier(max_leaf_nodes=10)
clf.fit(X_train, y_train)
predict = clf.predict(X_test)
print(accuracy_score(y_test, predict))
```

```python
from sklearn import tree
fig = plt.figure(figsize = (10, 7))
tree.plot_tree(clf, feature_names = X.columns,
              class_names = ["APROBADO","DESAPROBADO"], filled = True)
plt.show()
```

#### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
```

```python
rf = RandomForestClassifier(n_estimators = 100)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
print("Exactitud del Modelo: ", metrics.accuracy_score(y_test, y_pred))
```
